```python
import safetensors

def load_learned_embeddings(text_encoder, tokenizer, placeholder_token_ids, embeddings_path):
    """
    Load learned embeddings from a .safetensors or .bin file
    
    Args:
        text_encoder: The CLIP text encoder
        tokenizer: The CLIP tokenizer
        placeholder_token_ids: List of token IDs for placeholder tokens
        embeddings_path: Path to the saved embeddings file (.safetensors or .bin)
    """
    # Load the embeddings
    if embeddings_path.endswith('.safetensors'):
        learned_embeds_dict = safetensors.torch.load_file(embeddings_path)
    else:
        learned_embeds_dict = torch.load(embeddings_path)
    
    # Get the learned embeddings tensor
    # The dict typically has one key (the placeholder token)
    learned_embeds = list(learned_embeds_dict.values())[0]
    
    # Get the current embeddings
    token_embeds = text_encoder.get_input_embeddings().weight.data
    
    # Replace placeholder token embeddings with learned ones
    with torch.no_grad():
        for i, token_id in enumerate(placeholder_token_ids):
            if i < learned_embeds.shape[0]:
                token_embeds[token_id] = learned_embeds[i].to(token_embeds.device, dtype=token_embeds.dtype)
    
    print(f"✓ Loaded learned embeddings from {embeddings_path}")
    print(f"  Shape: {learned_embeds.shape}")
    print(f"  Applied to {len(placeholder_token_ids)} token(s)")
```

```python
class TextualInversionDataset(Dataset):
    def __init__(
        self,
        data_root,
        tokenizer_1,
        tokenizer_2,
        size=1024,
        repeats=100,
        interpolation="bicubic",
        flip_p=0.5,
        set="train",
        placeholder_token="*",
        learnable_property="object",
        center_crop=False,
    ):
        self.data_root = data_root
        self.tokenizer_1 = tokenizer_1
        self.tokenizer_2 = tokenizer_2
        self.size = size
        self.placeholder_token = placeholder_token
        self.learnable_property = learnable_property
        self.center_crop = center_crop
        self.flip_p = flip_p

        # Load images and captions
        self.image_paths = []
        self.captions = []
        
        if not os.path.exists(self.data_root):
             raise ValueError(f"Data root {self.data_root} does not exist")

        for file_path in os.listdir(self.data_root):
            if file_path.lower().endswith(('.png', '.jpg', '.jpeg', '.webp')):
                img_path = os.path.join(self.data_root, file_path)
                self.image_paths.append(img_path)
                
                # Look for caption file (same name with .txt extension)
                caption_path = os.path.splitext(img_path)[0] + '.txt'
                if os.path.exists(caption_path):
                    with open(caption_path, 'r') as f:
                        caption = f.read().strip()
                else:
                    caption = ""  # Default empty caption
                self.captions.append(caption)

        self.num_images = len(self.image_paths)
        self._length = self.num_images

        if set == "train":
            self._length = self.num_images * repeats
            
        # Standard interpolation map
        self.interpolation = {
            "linear": PIL.Image.Resampling.BILINEAR,
            "bilinear": PIL.Image.Resampling.BILINEAR,
            "bicubic": PIL.Image.Resampling.BICUBIC,
            "lanczos": PIL.Image.Resampling.LANCZOS,
            "nearest": PIL.Image.Resampling.NEAREST,
        }.get(interpolation, PIL.Image.Resampling.BICUBIC)

        self.flip_transform = transforms.RandomHorizontalFlip(p=self.flip_p)
        # We use a custom crop logic for SDXL to get coordinates

    def __len__(self):
        return self._length

    def __getitem__(self, i):
        example = {}
        idx = i % self.num_images
        image = Image.open(self.image_paths[idx])

        if not image.mode == "RGB":
            image = image.convert("RGB")
            
        # Store original size for SDXL micro-conditioning
        example["original_size"] = (image.height, image.width)

        # 1. Resize
        image = image.resize((self.size, self.size), resample=self.interpolation)

        # 2. Crop logic (Center or Random)
        if self.center_crop:
            y1 = max(0, int(round((image.height - self.size) / 2.0)))
            x1 = max(0, int(round((image.width - self.size) / 2.0)))
            image = TF.center_crop(image, (self.size, self.size))
        else:
            # Random crop logic
            # Since we resized to (size, size) above, random crop is trivial if it matches size
            # But usually SDXL logic resizes to *at least* size. 
            # For simplicity here, we assume resize was exact so crop is 0,0
            # If you want true random crop, resize should be larger than self.size
            y1, x1 = 0, 0 
            
        example["crop_top_left"] = (y1, x1)

        # 3. Flip
        # Note: If we flip, we technically change the "crop" relation, but usually ignored for simple TI
        # Or you can skip flip for SDXL to be safe with coordinates
        image = self.flip_transform(image)
        
        # 4. Prepare Text
        caption = self.captions[idx]
        if self.learnable_property == "object":
            if caption:
                text = f"a cropped image of {self.placeholder_token}: source_cartoon, bull anthro, {caption}"
            else:
                text = f"a cropped image of {self.placeholder_token}"
        else:
            if caption:
                text = f"A photo in the style of {self.placeholder_token}: {caption}"
            else:
                text = f"A photo in the style of {self.placeholder_token}"

        # Tokenize for BOTH tokenizers
        example["input_ids_1"] = self.tokenizer_1(
            text,
            padding="max_length",
            truncation=True,
            max_length=self.tokenizer_1.model_max_length,
            return_tensors="pt",
        ).input_ids[0]

        example["input_ids_2"] = self.tokenizer_2(
            text,
            padding="max_length",
            truncation=True,
            max_length=self.tokenizer_2.model_max_length,
            return_tensors="pt",
        ).input_ids[0]

        # 5. Normalize Image
        img = np.array(image).astype(np.uint8)
        image = (img / 127.5 - 1.0).astype(np.float32)

        example["pixel_values"] = torch.from_numpy(image).permute(2, 0, 1)
        return example
```

```python

def save_model_card(repo_id: str, images: list = None, base_model: str = None, repo_folder: str = None):
    img_str = ""
    if images is not None:
        for i, image in enumerate(images):
            image.save(os.path.join(repo_folder, f"image_{i}.png"))
            img_str += f"![img_{i}](./image_{i}.png)\n"
    model_description = f"""
# Textual inversion text2image fine-tuning - {repo_id}
These are textual inversion adaption weights for {base_model}. You can find some example images in the following. \n
{img_str}
"""
    model_card = load_or_create_model_card(
        repo_id_or_path=repo_id,
        from_training=True,
        license="creativeml-openrail-m",
        base_model=base_model,
        model_description=model_description,
        inference=True,
    )

    tags = [
        "stable-diffusion",
        "stable-diffusion-diffusers",
        "text-to-image",
        "diffusers",
        "textual_inversion",
        "diffusers-training",
    ]
    model_card = populate_model_card(model_card, tags=tags)

    model_card.save(os.path.join(repo_folder, "README.md"))


def save_progress(text_encoder, placeholder_token_ids, accelerator, args, save_path, safe_serialization=True):
    logger = get_logger(__name__)
    logger.info(f"Saving embeddings to {save_path}")
    learned_embeds = (
        accelerator.unwrap_model(text_encoder)
        .get_input_embeddings()
        .weight[min(placeholder_token_ids) : max(placeholder_token_ids) + 1]
    )
    learned_embeds_dict = {args.placeholder_token: learned_embeds.detach().cpu()}

    if safe_serialization:
        safetensors.torch.save_file(learned_embeds_dict, save_path, metadata={"format": "pt"})
    else:
        torch.save(learned_embeds_dict, save_path)

def log_validation(text_encoder_1, text_encoder_2, tokenizer_1, tokenizer_2, unet, vae, args, accelerator, weight_dtype):
    logger = get_logger(__name__)
    logger.info(
        f"Running validation... \n Generating {args.num_validation_images} images with prompt:"
        f" {args.validation_prompt}."
    )
    
    # Move VAE to GPU for validation
    vae.to(accelerator.device)
    
    pipeline = StableDiffusionXLPipeline.from_pretrained(
        args.pretrained_model_name_or_path,
        text_encoder=accelerator.unwrap_model(text_encoder_1),
        text_encoder_2=accelerator.unwrap_model(text_encoder_2),
        tokenizer=tokenizer_1,
        tokenizer_2=tokenizer_2,
        unet=unet,
        vae=vae,
        revision=args.revision,
        variant=args.variant,
        torch_dtype=weight_dtype,
    )
    
    pipeline.scheduler = DPMSolverMultistepScheduler.from_config(pipeline.scheduler.config)
    pipeline = pipeline.to(accelerator.device)
    pipeline.set_progress_bar_config(disable=True)

    # run inference
    generator = None if args.seed is None else torch.Generator(device=accelerator.device).manual_seed(args.seed)
    images = []
    
    if torch.backends.mps.is_available():
        autocast_ctx = nullcontext()
    else:
        autocast_ctx = torch.autocast(accelerator.device.type)

    with autocast_ctx:
        for _ in range(args.num_validation_images):
            image = pipeline(args.validation_prompt, num_inference_steps=25, generator=generator).images[0]
            images.append(image)

    for tracker in accelerator.trackers:
        if tracker.name == "wandb":
            tracker.log(
                {
                    "validation": [
                        wandb.Image(image, caption=f"{i}: {args.validation_prompt}") for i, image in enumerate(images)
                    ]
                }
            )

    del pipeline
    
    # Move VAE back to CPU to save memory
    vae.to("cpu")
    torch.cuda.empty_cache()
    
    return images
```

```python
def compute_token_similarities(text_encoder, tokenizer, placeholder_token_ids, reference_tokens):
    """
    Compute cosine similarity between learned token embeddings and reference tokens.
    
    Args:
        text_encoder: The CLIP text encoder
        tokenizer: The CLIP tokenizer
        placeholder_token_ids: List of token IDs for the learned placeholder tokens
        reference_tokens: List of token strings to compare against
    
    Returns:
        Dictionary mapping reference token names to similarity scores
    """
    embeddings = text_encoder.get_input_embeddings().weight
    learned_embedding = embeddings[placeholder_token_ids].mean(dim=0)
    learned_embedding_norm = F.normalize(learned_embedding.unsqueeze(0), p=2, dim=1)
    
    similarities = {}
    
    for token_str in reference_tokens:
        token_ids = tokenizer.encode(token_str, add_special_tokens=False)
        
        if len(token_ids) == 0:
            continue
            
        ref_embeddings = embeddings[token_ids]
        ref_embedding = ref_embeddings.mean(dim=0)
        ref_embedding_norm = F.normalize(ref_embedding.unsqueeze(0), p=2, dim=1)
        
        similarity = F.cosine_similarity(learned_embedding_norm, ref_embedding_norm).item()
        similarities[token_str] = similarity
    
    return similarities
```

```python
class TextualInversionTrainerSDXL:
    def __init__(self, config: "TrainingConfig"):
        self.config = config
        self.accelerator = None
        self.tokenizer_1 = None
        self.tokenizer_2 = None
        self.text_encoder_1 = None
        self.text_encoder_2 = None
        self.vae = None
        self.unet = None
        self.noise_scheduler = None
        self.optimizer = None
        self.lr_scheduler = None
        self.train_dataloader = None
        self.placeholder_token_ids_1 = None
        self.placeholder_token_ids_2 = None
        self.weight_dtype = None
        self.repo_id = None

    def setup(self):
        """Initialize all components for SDXL"""
        args = self.config
        logger = get_logger(__name__)
        
        # Setup logging
        logging.basicConfig(
            format="%(asctime)s - %(levelname)s - %(name)s - %(message)s",
            datefmt="%m/%d/%Y %H:%M:%S",
            level=logging.INFO,
        )
        
        # Setup accelerator
        logging_dir = os.path.join(args.output_dir, args.logging_dir)
        accelerator_project_config = ProjectConfiguration(project_dir=args.output_dir, logging_dir=logging_dir)
        self.accelerator = Accelerator(
            gradient_accumulation_steps=args.gradient_accumulation_steps,
            mixed_precision=args.mixed_precision,
            log_with=args.report_to,
            project_config=accelerator_project_config,
        )

        if self.accelerator.is_main_process:
            if args.output_dir is not None:
                os.makedirs(args.output_dir, exist_ok=True)
            if args.push_to_hub:
                self.repo_id = create_repo(
                    repo_id=args.hub_model_id or Path(args.output_dir).name, 
                    exist_ok=True, 
                    token=args.hub_token
                ).repo_id

        # Load Tokenizers
        self.tokenizer_1 = CLIPTokenizer.from_pretrained(args.pretrained_model_name_or_path, subfolder="tokenizer")
        self.tokenizer_2 = CLIPTokenizer.from_pretrained(args.pretrained_model_name_or_path, subfolder="tokenizer_2")

        # Load Schedulers and Models
        self.noise_scheduler = DDPMScheduler.from_pretrained(args.pretrained_model_name_or_path, subfolder="scheduler")
        self.text_encoder_1 = CLIPTextModel.from_pretrained(
            args.pretrained_model_name_or_path, subfolder="text_encoder", revision=args.revision
        )
        self.text_encoder_2 = CLIPTextModelWithProjection.from_pretrained(
            args.pretrained_model_name_or_path, subfolder="text_encoder_2", revision=args.revision
        )
        self.vae = AutoencoderKL.from_pretrained(
            args.pretrained_model_name_or_path, subfolder="vae", revision=args.revision, variant=args.variant
        )
        self.unet = UNet2DConditionModel.from_pretrained(
            args.pretrained_model_name_or_path, subfolder="unet", revision=args.revision, variant=args.variant
        )
        
        # OOM Fix: Enable VAE optimizations
        self.vae.enable_slicing()
        self.vae.enable_tiling()

        # Handle Placeholders for BOTH tokenizers
        placeholder_tokens = [args.placeholder_token]
        if args.num_vectors > 1:
            additional_tokens = [f"{args.placeholder_token}_{i}" for i in range(1, args.num_vectors)]
            placeholder_tokens += additional_tokens

        # Add tokens to tokenizer 1
        num_added_tokens = self.tokenizer_1.add_tokens(placeholder_tokens)
        if num_added_tokens != args.num_vectors:
            raise ValueError(f"The tokenizer 1 already contains the token {args.placeholder_token}.")
            
        # Add tokens to tokenizer 2
        num_added_tokens_2 = self.tokenizer_2.add_tokens(placeholder_tokens)
        if num_added_tokens_2 != args.num_vectors:
            raise ValueError(f"The tokenizer 2 already contains the token {args.placeholder_token}.")

        # Get IDs
        self.placeholder_token_ids_1 = self.tokenizer_1.convert_tokens_to_ids(placeholder_tokens)
        self.placeholder_token_ids_2 = self.tokenizer_2.convert_tokens_to_ids(placeholder_tokens)

        # Resize embeddings
        self.text_encoder_1.resize_token_embeddings(len(self.tokenizer_1))
        self.text_encoder_2.resize_token_embeddings(len(self.tokenizer_2))

        # Initialization Logic
        token_embeds_1 = self.text_encoder_1.get_input_embeddings().weight.data
        token_embeds_2 = self.text_encoder_2.get_input_embeddings().weight.data

        # Get initializer token IDs
        init_token_ids_1 = self.tokenizer_1.encode(args.initializer_token, add_special_tokens=False)
        init_token_ids_2 = self.tokenizer_2.encode(args.initializer_token, add_special_tokens=False)
        
        # Simple initialization strategy: use first token found
        if len(init_token_ids_1) > 0 and len(init_token_ids_2) > 0:
            init_id_1 = init_token_ids_1[0]
            init_id_2 = init_token_ids_2[0]
            
            with torch.no_grad():
                for i, token_id in enumerate(self.placeholder_token_ids_1):
                    token_embeds_1[token_id] = token_embeds_1[init_id_1].clone()
                    # Add small noise if multi-vector
                    if i > 0: 
                        token_embeds_1[token_id] += torch.randn_like(token_embeds_1[token_id]) * 0.01

                for i, token_id in enumerate(self.placeholder_token_ids_2):
                    token_embeds_2[token_id] = token_embeds_2[init_id_2].clone()
                    if i > 0: 
                        token_embeds_2[token_id] += torch.randn_like(token_embeds_2[token_id]) * 0.01
            
            logger.info(f"Initialized from {args.initializer_token}")
        else:
            logger.warning("Initializer token not found, using random initialization")

        # Freeze models
        self.vae.requires_grad_(False)
        self.unet.requires_grad_(False)
        self.text_encoder_1.text_model.encoder.requires_grad_(False)
        self.text_encoder_1.text_model.final_layer_norm.requires_grad_(False)
        self.text_encoder_1.text_model.embeddings.position_embedding.requires_grad_(False)
        self.text_encoder_2.text_model.encoder.requires_grad_(False)
        self.text_encoder_2.text_model.final_layer_norm.requires_grad_(False)
        self.text_encoder_2.text_model.embeddings.position_embedding.requires_grad_(False)

        # Gradient Checkpointing
        if args.gradient_checkpointing:
            self.text_encoder_1.gradient_checkpointing_enable()
            self.text_encoder_2.gradient_checkpointing_enable()
            self.unet.enable_gradient_checkpointing()

        # Optimizer: optimize embedding layers of BOTH encoders
        self.optimizer = torch.optim.AdamW(
            [
                self.text_encoder_1.get_input_embeddings().weight,
                self.text_encoder_2.get_input_embeddings().weight
            ],
            lr=args.learning_rate,
            betas=(args.adam_beta1, args.adam_beta2),
            weight_decay=args.adam_weight_decay,
            eps=args.adam_epsilon,
        )

        # Dataset
        train_dataset = TextualInversionDataset(
            data_root=args.train_data_dir,
            tokenizer_1=self.tokenizer_1,
            tokenizer_2=self.tokenizer_2,
            size=args.resolution,
            placeholder_token=(" ".join(self.tokenizer_1.convert_ids_to_tokens(self.placeholder_token_ids_1))),
            repeats=args.repeats,
            center_crop=args.center_crop,
            set="train",
        )
        
        self.train_dataloader = torch.utils.data.DataLoader(
            train_dataset, 
            batch_size=args.train_batch_size, 
            shuffle=True, 
            num_workers=args.dataloader_num_workers
        )

        # Scheduler
        num_update_steps_per_epoch = math.ceil(len(self.train_dataloader) / args.gradient_accumulation_steps)
        if args.max_train_steps is None:
            args.max_train_steps = args.num_train_epochs * num_update_steps_per_epoch
        
        self.lr_scheduler = get_scheduler(
            args.lr_scheduler,
            optimizer=self.optimizer,
            num_warmup_steps=args.lr_warmup_steps * self.accelerator.num_processes,
            num_training_steps=args.max_train_steps * self.accelerator.num_processes,
            num_cycles=args.lr_num_cycles,
        )

        # Prepare
        self.text_encoder_1.train()
        self.text_encoder_2.train()
        
        (
            self.text_encoder_1,
            self.text_encoder_2,
            self.optimizer,
            self.train_dataloader,
            self.lr_scheduler
        ) = self.accelerator.prepare(
            self.text_encoder_1, 
            self.text_encoder_2, 
            self.optimizer, 
            self.train_dataloader, 
            self.lr_scheduler
        )

        # Cast non-trainables to weight_dtype
        weight_dtype = torch.float32
        if self.accelerator.mixed_precision == "fp16":
            weight_dtype = torch.float16
        elif self.accelerator.mixed_precision == "bf16":
            weight_dtype = torch.bfloat16
        self.weight_dtype = weight_dtype

        self.unet.to(self.accelerator.device, dtype=weight_dtype)
        # We don't move VAE to GPU yet, we do it in precompute step or validation
        
        # Initialize trackers
        if self.accelerator.is_main_process:
            self.accelerator.init_trackers(args.wandb_project_name, config=vars(args))

        return num_update_steps_per_epoch

    def log_token_similarities(self, global_step):
        """Simple check on first text encoder"""
        logger = get_logger(__name__)
        if not self.accelerator.is_main_process:
            return None
            
        # Just check Encoder 1 for simplicity
        embeddings = self.accelerator.unwrap_model(self.text_encoder_1).get_input_embeddings().weight
        learned_embed = embeddings[self.placeholder_token_ids_1].detach().mean(dim=0)
        norm_learned = F.normalize(learned_embed.unsqueeze(0), p=2, dim=1)
        
        ref_tokens = ["bull", "cartoon", "character"]
        
        logger.info(f"\nStep {global_step} - Similarities:")
        metrics = {}
        for token in ref_tokens:
            tid = self.tokenizer_1.encode(token, add_special_tokens=False)[0]
            ref_embed = embeddings[tid].detach()
            norm_ref = F.normalize(ref_embed.unsqueeze(0), p=2, dim=1)
            sim = F.cosine_similarity(norm_learned, norm_ref).item()
            logger.info(f"  {token}: {sim:.4f}")
            metrics[f"sim/{token}"] = sim
            
        return metrics

    def precompute_latents(self):
        """Pre-compute and cache all latents to save memory during training"""
        args = self.config
        logger = get_logger(__name__)
        
        logger.info("Pre-computing latents to save memory...")
        cached_samples = []
        
        # Move VAE to GPU for encoding
        self.vae.to(self.accelerator.device, dtype=self.weight_dtype)
        
        # Create a progress bar for caching
        cache_progress = tqdm(
            self.train_dataloader, 
            desc="Caching Latents",
            disable=not self.accelerator.is_local_main_process
        )
        
        for batch in cache_progress:
            with torch.no_grad():
                # Encode latents
                pixel_values = batch["pixel_values"].to(self.accelerator.device, dtype=self.weight_dtype)
                latents = self.vae.encode(pixel_values).latent_dist.sample()
                latents = latents * self.vae.config.scaling_factor
                
                # Split batch into individual samples for proper shuffling later
                bsz = latents.shape[0]
                for i in range(bsz):
                    sample = {
                        "latents": latents[i].detach().cpu(),
                        "input_ids_1": batch["input_ids_1"][i],
                        "input_ids_2": batch["input_ids_2"][i],
                        "original_size": batch["original_size"][i], 
                        "crop_top_left": batch["crop_top_left"][i]
                    }
                    cached_samples.append(sample)

        # Validate we have samples
        if len(cached_samples) == 0:
            raise ValueError("No samples were cached! Check your dataset.")
        
        # CRITICAL: Move VAE to CPU to free ~3GB VRAM
        logger.info(f"Cached {len(cached_samples)} samples. Unloading VAE from GPU...")
        torch.cuda.synchronize()
        self.vae.to("cpu")
        torch.cuda.empty_cache()
        gc.collect()
        
        return cached_samples

    def create_batch_iterator(self, cached_samples, batch_size):
        """Create iterator that yields batches from cached samples"""
        # Shuffle samples at the start
        indices = list(range(len(cached_samples)))
        random.shuffle(indices)
        
        for i in range(0, len(cached_samples), batch_size):
            batch_indices = indices[i:i + batch_size]
            batch_samples = [cached_samples[idx] for idx in batch_indices]
            
            # Collate on the fly
            batch = {
                "latents": torch.stack([s["latents"] for s in batch_samples]),
                "input_ids_1": torch.stack([s["input_ids_1"] for s in batch_samples]),
                "input_ids_2": torch.stack([s["input_ids_2"] for s in batch_samples]),
                "original_size": torch.stack([s["original_size"] for s in batch_samples]),
                "crop_top_left": torch.stack([s["crop_top_left"] for s in batch_samples])
            }
            yield batch

    def train_step(self, batch, orig_embeds_1, orig_embeds_2):
        """Single training step"""
        args = self.config
        
        # 1. Get Latents (Move back to GPU with correct dtype)
        latents = batch["latents"].to(self.accelerator.device, dtype=self.weight_dtype)

        # 2. Sample noise
        noise = torch.randn_like(latents)
        bsz = latents.shape[0]
        timesteps = torch.randint(
            0, 
            self.noise_scheduler.config.num_train_timesteps, 
            (bsz,), 
            device=latents.device
        ).long()
        noisy_latents = self.noise_scheduler.add_noise(latents, noise, timesteps)

        # 3. Get text embeddings from BOTH encoders
        # Encoder 1
        enc_out_1 = self.text_encoder_1(
            batch["input_ids_1"].to(self.accelerator.device), 
            output_hidden_states=True
        )
        hidden_states_1 = enc_out_1.hidden_states[-2].to(dtype=self.weight_dtype) 
        
        # Encoder 2
        enc_out_2 = self.text_encoder_2(
            batch["input_ids_2"].to(self.accelerator.device), 
            output_hidden_states=True
        )
        hidden_states_2 = enc_out_2.hidden_states[-2].to(dtype=self.weight_dtype)
        pooled_embeds_2 = enc_out_2.text_embeds.to(dtype=self.weight_dtype)
        
        # Concatenate hidden states
        encoder_hidden_states = torch.cat([hidden_states_1, hidden_states_2], dim=-1)

        # 4. Prepare Added Conditions (Micro-Conditioning)
        def compute_time_ids(original_size, crop_coords):
            target_size = [args.resolution, args.resolution]
            return list(original_size + crop_coords + target_size)
        
        add_time_ids = [
            compute_time_ids(
                batch["original_size"][i].tolist(),
                batch["crop_top_left"][i].tolist()
            )
            for i in range(bsz)
        ]
        add_time_ids = torch.tensor(add_time_ids).to(
            self.accelerator.device, 
            dtype=self.weight_dtype
        )

        added_cond_kwargs = {
            "text_embeds": pooled_embeds_2,
            "time_ids": add_time_ids
        }

        # 5. Predict
        model_pred = self.unet(
            noisy_latents, 
            timesteps, 
            encoder_hidden_states, 
            added_cond_kwargs=added_cond_kwargs
        ).sample

        # 6. Loss
        if self.noise_scheduler.config.prediction_type == "epsilon":
            target = noise
        elif self.noise_scheduler.config.prediction_type == "v_prediction":
            target = self.noise_scheduler.get_velocity(latents, noise, timesteps)
        else:
            raise ValueError(f"Unknown prediction type: {self.noise_scheduler.config.prediction_type}")

        loss = F.mse_loss(model_pred.float(), target.float(), reduction="mean")

        # 7. Backprop
        self.accelerator.backward(loss)
        self.optimizer.step()
        self.lr_scheduler.step()
        self.optimizer.zero_grad()

        # 8. Restore non-placeholder tokens
        # Encoder 1
        index_no_updates_1 = torch.ones((len(self.tokenizer_1),), dtype=torch.bool)
        index_no_updates_1[min(self.placeholder_token_ids_1) : max(self.placeholder_token_ids_1) + 1] = False
        with torch.no_grad():
            self.accelerator.unwrap_model(self.text_encoder_1).get_input_embeddings().weight[
                index_no_updates_1
            ] = orig_embeds_1[index_no_updates_1]
        
        # Encoder 2
        index_no_updates_2 = torch.ones((len(self.tokenizer_2),), dtype=torch.bool)
        index_no_updates_2[min(self.placeholder_token_ids_2) : max(self.placeholder_token_ids_2) + 1] = False
        with torch.no_grad():
            self.accelerator.unwrap_model(self.text_encoder_2).get_input_embeddings().weight[
                index_no_updates_2
            ] = orig_embeds_2[index_no_updates_2]
        
        return loss.detach().item()

    def train(self):
        args = self.config
        num_update_steps_per_epoch = self.setup()
        logger = get_logger(__name__)
        
        logger.info("***** Running training SDXL *****")
        logger.info(f"  Num examples = {len(self.train_dataloader.dataset)}")
        logger.info(f"  Num Epochs = {args.num_train_epochs}")
        logger.info(f"  Instantaneous batch size per device = {args.train_batch_size}")
        logger.info(f"  Gradient Accumulation steps = {args.gradient_accumulation_steps}")
        logger.info(f"  Total optimization steps = {args.max_train_steps}")
        logger.info(f"  Save every {args.checkpointing_steps} steps")  # ADD THIS
        
        global_step = 0
        first_epoch = 0
        
        # Store original embeddings for restoration
        orig_embeds_1 = self.accelerator.unwrap_model(
            self.text_encoder_1
        ).get_input_embeddings().weight.data.clone()
        orig_embeds_2 = self.accelerator.unwrap_model(
            self.text_encoder_2
        ).get_input_embeddings().weight.data.clone()

        # Pre-compute latents
        cached_samples = self.precompute_latents()
        
        # Calculate batches per epoch
        batches_per_epoch = len(cached_samples) // args.train_batch_size
        logger.info(f"  Batches per epoch = {batches_per_epoch}")

        # Main progress bar
        progress_bar = tqdm(
            range(args.max_train_steps), 
            disable=not self.accelerator.is_local_main_process,
            desc="Training Steps"
        )

        # ADD: Helper function to save checkpoint
        def save_checkpoint(step, is_final=False):
            if not self.accelerator.is_main_process:
                return
                
            suffix = "final" if is_final else f"steps-{step}"
            
            # Save encoder 1
            save_path_1 = os.path.join(
                args.output_dir, 
                f"learned_embeds-{suffix}.safetensors"
            )
            save_progress(
                self.text_encoder_1, 
                self.placeholder_token_ids_1, 
                self.accelerator, 
                args, 
                save_path_1
            )
            
            # Save encoder 2
            save_path_2 = os.path.join(
                args.output_dir, 
                f"learned_embeds_2-{suffix}.safetensors"
            )
            save_progress(
                self.text_encoder_2, 
                self.placeholder_token_ids_2, 
                self.accelerator, 
                args, 
                save_path_2
            )
            
            logger.info(f"✓ Saved checkpoint at step {step} (is_final={is_final})")

        # Training Loop
        try:  # ADD: Wrap in try-except to catch interruptions
            for epoch in range(first_epoch, args.num_train_epochs):
                self.text_encoder_1.train()
                self.text_encoder_2.train()
                
                # Create batch iterator for this epoch
                batch_iterator = self.create_batch_iterator(cached_samples, args.train_batch_size)
                
                for batch in batch_iterator:
                    # Check if we've reached max steps
                    if global_step >= args.max_train_steps:
                        break
                        
                    with self.accelerator.accumulate([self.text_encoder_1, self.text_encoder_2]):
                        loss = self.train_step(batch, orig_embeds_1, orig_embeds_2)

                    # Logging and checkpointing (only when gradients sync)
                    if self.accelerator.sync_gradients:
                        progress_bar.update(1)
                        global_step += 1
                        
                        # Prepare logs
                        logs = {
                            "loss": loss, 
                            "lr": self.lr_scheduler.get_last_lr()[0],
                            "epoch": epoch,
                            "global_step": global_step  # ADD: Track actual step
                        }
                        
                        # FIXED: More frequent checkpointing
                        should_save = (global_step % args.checkpointing_steps == 0) or (global_step == args.max_train_steps)
                        
                        if should_save:
                            logger.info(f"\n{'='*50}")
                            logger.info(f"CHECKPOINT at step {global_step}/{args.max_train_steps}")
                            logger.info(f"{'='*50}")
                            
                            # Save checkpoint
                            save_checkpoint(global_step, is_final=(global_step == args.max_train_steps))

                            # Run validation
                            if args.validation_prompt:
                                logger.info("Running validation...")
                                log_validation(
                                    self.text_encoder_1, 
                                    self.text_encoder_2,
                                    self.tokenizer_1, 
                                    self.tokenizer_2,
                                    self.unet, 
                                    self.vae,
                                    args, 
                                    self.accelerator, 
                                    self.weight_dtype, 
                                )
                                
                            # Log token similarities
                            sim_metrics = self.log_token_similarities(global_step)
                            if sim_metrics: 
                                logs.update(sim_metrics)
                            
                        progress_bar.set_postfix(**logs)
                        self.accelerator.log(logs, step=global_step)

                # Check if we've reached max steps at epoch end
                if global_step >= args.max_train_steps:
                    break

        except KeyboardInterrupt:
            logger.warning("\n⚠️ Training interrupted! Saving emergency checkpoint...")
            save_checkpoint(global_step, is_final=False)
            logger.info("Emergency checkpoint saved. You can resume training.")
            raise
        
        except Exception as e:
            logger.error(f"\n❌ Training failed with error: {e}")
            logger.warning("Attempting to save emergency checkpoint...")
            try:
                save_checkpoint(global_step, is_final=False)
                logger.info("Emergency checkpoint saved.")
            except:
                logger.error("Failed to save emergency checkpoint!")
            raise

        # End of training
        self.accelerator.wait_for_everyone()
        
        if self.accelerator.is_main_process:
            logger.info("\n" + "="*60)
            logger.info("TRAINING COMPLETED - Saving final checkpoint...")
            logger.info("="*60)
            
            # Final Save - Use the helper function
            save_checkpoint(global_step, is_final=True)
            
            # Final validation
            if args.validation_prompt:
                logger.info("Running final validation...")
                log_validation(
                    self.text_encoder_1, 
                    self.text_encoder_2,
                    self.tokenizer_1, 
                    self.tokenizer_2,
                    self.unet, 
                    self.vae,
                    args, 
                    self.accelerator, 
                    self.weight_dtype, 
                    args.num_train_epochs - 1,
                    global_step
                )

            # Push to hub if requested
            if args.push_to_hub:
                logger.info("Pushing to hub...")
                upload_folder(
                    repo_id=self.repo_id,
                    folder_path=args.output_dir,
                    commit_message="End of training",
                    ignore_patterns=["step_*", "epoch_*"],
                )

        self.accelerator.end_training()
        logger.info("✓ Training completed successfully!")
```

```python
def train_textual_inversion(
    train_data_dir: str,
    placeholder_token: str = "<my-token>",
    initializer_token: str = "a",
    learnable_property: str = "object",
    output_dir: str = "textual-inversion-output",
    pretrained_model_name_or_path: str = "John6666/autismmix-sdxl-autismmix-pony-sdxl",
    resolution: int = 512,
    train_batch_size: int = 4,
    learning_rate: float = 5e-4,
    max_train_steps: int = 5000,
    save_steps: int = 500,
    validation_prompt: str = None,
    validation_steps: int = 100,
    num_validation_images: int = 4,
    wandb_project_name: str = "textual-inversion",
    seed: int = 42,
    mixed_precision: str = "no",
    gradient_accumulation_steps: int = 1,
    num_vectors: int = 1,
    repeats: int = 100,
    center_crop: bool = False,
    lr_scheduler: str = "constant",
    lr_warmup_steps: int = 500,
    checkpointing_steps: int = 500,
    resume_from_checkpoint: str = None,
    pretrained_embeddings_path=None,
):
    """
    Train a textual inversion model.
    
    Args:
        train_data_dir: Path to training data directory containing images and .txt caption files
        placeholder_token: Token to learn (e.g., "<my-cat>")
        initializer_token: Token to initialize embedding (e.g., "cat", "dog", "object", "style")
        output_dir: Where to save the trained embeddings
        pretrained_model_name_or_path: Base Stable Diffusion model
        resolution: Training image resolution
        train_batch_size: Batch size per device
        learning_rate: Learning rate
        max_train_steps: Maximum training steps
        save_steps: Save embeddings every N steps
        validation_prompt: Prompt for validation (should include placeholder_token)
        validation_steps: Run validation every N steps
        num_validation_images: Number of validation images to generate
        wandb_project_name: Weights & Biases project name
        seed: Random seed
        mixed_precision: "no", "fp16", or "bf16"
        gradient_accumulation_steps: Gradient accumulation steps
        num_vectors: Number of vectors to learn
        repeats: How many times to repeat the training data
        center_crop: Whether to center crop images
        lr_scheduler: Learning rate scheduler type
        lr_warmup_steps: Warmup steps for learning rate
        checkpointing_steps: Save checkpoint every N steps
        resume_from_checkpoint: Path to checkpoint to resume from
    """
    
    config = TrainingConfig(
        train_data_dir=train_data_dir,
        placeholder_token=placeholder_token,
        initializer_token=initializer_token,
        learnable_property=learnable_property,
        output_dir=output_dir,
        pretrained_model_name_or_path=pretrained_model_name_or_path,
        resolution=resolution,
        train_batch_size=train_batch_size,
        learning_rate=learning_rate,
        max_train_steps=max_train_steps,
        save_steps=save_steps,
        validation_prompt=validation_prompt,
        validation_steps=validation_steps,
        num_validation_images=num_validation_images,
        wandb_project_name=wandb_project_name,
        seed=seed,
        mixed_precision=mixed_precision,
        gradient_accumulation_steps=gradient_accumulation_steps,
        num_vectors=num_vectors,
        repeats=repeats,
        center_crop=center_crop,
        lr_scheduler=lr_scheduler,
        lr_warmup_steps=lr_warmup_steps,
        checkpointing_steps=checkpointing_steps,
        resume_from_checkpoint=resume_from_checkpoint,
        report_to="wandb",
        pretrained_embeddings_path=pretrained_embeddings_path,
    )
    
    trainer = TextualInversionTrainerSDXL(config)
    trainer.train()
    
    return trainer
```
