```python
import logging
import math
import os
import random
from pathlib import Path
from dataclasses import dataclass
from PIL import Image

import torch
import torch.nn.functional as F
import torchvision.transforms.functional as TF
from torch.utils.data import Dataset
from torchvision import transforms
from accelerate import Accelerator
from accelerate.utils import ProjectConfiguration, set_seed
from diffusers import (
    AutoencoderKL,
    DDPMScheduler,
    UNet2DConditionModel,
    StableDiffusionPipeline
)
from diffusers.optimization import get_scheduler
from peft import LoraConfig, get_peft_model, PeftModel
from tqdm.auto import tqdm
from transformers import CLIPTextModel, CLIPTokenizer
from torchmetrics.multimodal import CLIPScore

logger = logging.getLogger(__name__)
```

```python
# --- 1. DATASET CLASS (Must be defined before Trainer) ---
class LoRADataset(Dataset):
    def __init__(
        self,
        data_root,
        tokenizer,
        size=512,
        placeholder_token="<Gau>",
        center_crop=True,
        flip_p=0.0, # Keep 0.0 for faces/characters usually
    ):
        self.data_root = data_root
        self.tokenizer = tokenizer
        self.size = size
        self.placeholder_token = placeholder_token
        self.center_crop = center_crop
        self.flip_p = flip_p

        self.image_paths = [os.path.join(data_root, file_path) for file_path in os.listdir(data_root) if file_path.endswith(('.png', '.jpg', '.jpeg'))]
        
        # Simple templates
        self.templates = [
            "a photo of {}",
            "a rendering of {}",
            "a close-up photo of a {}",
            "a photo of a clean {}",
            "a dark photo of the {}",
            "a photo of my {}",
        ]

        self.image_transforms = transforms.Compose(
            [
                transforms.Resize(size, interpolation=transforms.InterpolationMode.BILINEAR),
                transforms.CenterCrop(size) if center_crop else transforms.RandomCrop(size),
                transforms.RandomHorizontalFlip(p=flip_p),
                transforms.ToTensor(),
                transforms.Normalize([0.5], [0.5]),
            ]
        )

    def __len__(self):
        return len(self.image_paths)

    def __getitem__(self, index):
        example = {}
        img_path = self.image_paths[index]
        image = Image.open(img_path)

        if not image.mode == "RGB":
            image = image.convert("RGB")

        # 1. Image
        example["pixel_values"] = self.image_transforms(image)

        # 2. Text
        # Check for .txt caption file
        txt_path = os.path.splitext(img_path)[0] + ".txt"
        if os.path.exists(txt_path):
            with open(txt_path, "r") as f:
                prompt = f"A Photo of {self.placeholder_token} " + f.read().strip()
        else:
            prompt = random.choice(self.templates).format(self.placeholder_token)

        # Tokenize
        example["input_ids"] = self.tokenizer(
            prompt,
            padding="max_length",
            truncation=True,
            max_length=self.tokenizer.model_max_length,
            return_tensors="pt",
        ).input_ids[0]

        return example
```

```python




# --- 2. TRAINER CLASS ---
class LoRATrainer:
    def __init__(self, config):
        self.config = config
        self.accelerator = None
        self.tokenizer = None
        self.text_encoder = None
        self.vae = None
        self.unet = None
        self.noise_scheduler = None
        self.optimizer = None
        self.lr_scheduler = None
        self.train_dataloader = None
        self.weight_dtype = None
        self.clip_score_metric = None 

    def setup(self):
        args = self.config
        
        logging.basicConfig(
            format="%(asctime)s - %(levelname)s - %(name)s - %(message)s",
            level=logging.INFO,
        )
        accelerator_project_config = ProjectConfiguration(project_dir=args.output_dir)
        self.accelerator = Accelerator(
            gradient_accumulation_steps=args.gradient_accumulation_steps,
            mixed_precision=args.mixed_precision,
            log_with=args.report_to,  # <--- Quan trọng: Báo cho nó biết dùng "wandb"
            project_config=accelerator_project_config,
        )
        
        # Khởi tạo WandB Project
        if self.accelerator.is_main_process:
            self.accelerator.init_trackers(
                project_name=args.wandb_project_name, 
                config=vars(args)
            )

        # Load Models
        self.tokenizer = CLIPTokenizer.from_pretrained(args.pretrained_model_name_or_path, subfolder="tokenizer")
        self.noise_scheduler = DDPMScheduler.from_pretrained(args.pretrained_model_name_or_path, subfolder="scheduler")
        self.text_encoder = CLIPTextModel.from_pretrained(args.pretrained_model_name_or_path, subfolder="text_encoder", torch_dtype=torch.float16)
        self.vae = AutoencoderKL.from_pretrained(args.pretrained_model_name_or_path, subfolder="vae", torch_dtype=torch.float16)
        self.unet = UNet2DConditionModel.from_pretrained(args.pretrained_model_name_or_path, subfolder="unet", torch_dtype=torch.float16)

        # Freeze Base Models
        self.vae.requires_grad_(False)
        self.text_encoder.requires_grad_(False)
        self.unet.requires_grad_(False)

        # Inject LoRA
        if args.resume_from_checkpoint:
            logger.info(f"Loading LoRA weights from {args.resume_from_checkpoint}")
            self.unet = PeftModel.from_pretrained(
                self.unet, 
                args.resume_from_checkpoint,
                is_trainable=True
            )
            logger.info("LoRA weights loaded successfully!")
        else:
            unet_lora_config = LoraConfig(
                r=args.rank,
                lora_alpha=args.rank,
                init_lora_weights="gaussian",
                target_modules=["to_k", "to_q", "to_v", "to_out.0"],
            )
            self.unet = get_peft_model(self.unet, unet_lora_config)
        
        self.unet.print_trainable_parameters()

        # Optimizer
        self.optimizer = torch.optim.AdamW(
            self.unet.parameters(),
            lr=args.learning_rate,
            weight_decay=args.adam_weight_decay,
            betas=(0.9,0.98), 
            eps=1e-6,
        )

        # Dataset
        train_dataset = LoRADataset(
            data_root=args.train_data_dir,
            tokenizer=self.tokenizer,
            size=args.resolution,
            placeholder_token=args.placeholder_token,
            center_crop=args.center_crop,
            flip_p=args.flip_p # No flip for consistent face training usually
        )
        self.train_dataloader = torch.utils.data.DataLoader(
            train_dataset, batch_size=args.train_batch_size, shuffle=True
        )

        # Scheduler
        self.lr_scheduler = get_scheduler(
            args.lr_scheduler,
            optimizer=self.optimizer,
            num_warmup_steps=args.lr_warmup_steps,
            num_training_steps=args.max_train_steps,
        )

        self.unet, self.optimizer, self.train_dataloader, self.lr_scheduler = self.accelerator.prepare(
            self.unet, self.optimizer, self.train_dataloader, self.lr_scheduler
        )
        
        self.text_encoder.to(self.accelerator.device, dtype=torch.float16)
        self.vae.to(self.accelerator.device, dtype=torch.float16)

        # Setup CLIPScore
        if self.accelerator.is_main_process:
            logger.info("Loading CLIPScore metric...")
            try:
                self.clip_score_metric = CLIPScore(model_name_or_path="openai/clip-vit-base-patch32").to(self.accelerator.device)
                logger.info("CLIPScore metric loaded successfully.")
            except Exception as e:
                logger.error(f"Failed to load CLIPScore: {e}")
                self.clip_score_metric = None

    def train(self):
        self.setup()
        args = self.config
        global_step = 0
        
        progress_bar = tqdm(range(args.max_train_steps), desc="LoRA Steps")
        
        for epoch in range(args.max_train_steps):
            self.unet.train()
            train_loss = 0.0
            
            for step, batch in enumerate(self.train_dataloader):
                with self.accelerator.accumulate(self.unet):
                    latents = self.vae.encode(batch["pixel_values"].to(dtype=torch.float16)).latent_dist.sample()
                    latents = latents * self.vae.config.scaling_factor

                    noise = torch.randn_like(latents)
                    bsz = latents.shape[0]
                    timesteps = torch.randint(0, self.noise_scheduler.config.num_train_timesteps, (bsz,), device=latents.device).long()
                    
                    noisy_latents = self.noise_scheduler.add_noise(latents, noise, timesteps)
                    encoder_hidden_states = self.text_encoder(batch["input_ids"])[0]
                    model_pred = self.unet(noisy_latents, timesteps, encoder_hidden_states).sample

                    if self.noise_scheduler.config.prediction_type == "epsilon":
                        target = noise
                    else:
                        target = self.noise_scheduler.get_velocity(latents, noise, timesteps)

                    loss = F.mse_loss(model_pred.float(), target.float(), reduction="mean")

                    self.accelerator.backward(loss)
                    
                    # ============ ADD GRADIENT CLIPPING ============
                    if self.accelerator.sync_gradients and hasattr(args, 'max_grad_norm'):
                        max_norm_val = args.max_grad_norm
                        if isinstance(max_norm_val, (tuple, list)):
                            max_norm_val = max_norm_val[0]
                        
                        self.accelerator.clip_grad_norm_(self.unet.parameters(), float(max_norm_val))
                    
                    self.optimizer.step()
                    self.lr_scheduler.step()
                    self.optimizer.zero_grad()
                
                train_loss += loss.detach().item()

                if self.accelerator.sync_gradients:
                    progress_bar.update(1)
                    global_step += 1
                    
                    # --- VALIDATION ---
                    current_clip_score = None
                    wandb_images = []
                    if self.accelerator.is_main_process and global_step % args.validation_steps == 0:
                        images = log_validation(
                            self.text_encoder, self.tokenizer, self.unet, self.vae, 
                            args, self.accelerator, torch.float16, epoch
                        )
                        if images:
                            for idx, img in enumerate(images):
                                wandb_images.append(
                                    wandb.Image(img, caption=f"Step {global_step} - Img {idx}")
                                )
                        
                        if images and self.clip_score_metric is not None:
                            try:
                                images_tensor = torch.stack([TF.to_tensor(img) for img in images])
                                images_tensor = (images_tensor * 255).clamp(0, 255).to(torch.uint8).to(self.accelerator.device)
                                prompts_list = [args.validation_prompt] * len(images)
                                
                                self.clip_score_metric.reset()
                                clip_score = self.clip_score_metric(images_tensor, prompts_list)
                                current_clip_score = clip_score.detach().item()
                                logger.info(f"Step {global_step}: CLIP Score = {current_clip_score:.2f}")
                            except Exception as e:
                                logger.error(f"Scoring error: {e}")

                    # Logging
                    avg_train_loss = train_loss / args.gradient_accumulation_steps
                    logs = {
                        "loss": avg_train_loss,
                        "lr": self.lr_scheduler.get_last_lr()[0],
                    }
                    if current_clip_score is not None:
                        logs["val_clip_score"] = current_clip_score

                    if wandb_images:
                        logs["validation_images"] = wandb_images
                        
                    self.accelerator.log(logs, step=global_step)
                    progress_bar.set_postfix(**logs)
                    train_loss = 0.0
                    
                    # Save Checkpoint
                    if global_step % args.save_steps == 0:
                        save_path = os.path.join(args.output_dir, f"lora-step-{global_step}")
                        self.unet.save_pretrained(save_path)

                if global_step >= args.max_train_steps:
                    break
        
        self.unet.save_pretrained(args.output_dir)
        logger.info("LoRA training completed!")

# --- 3. HELPER FUNCTION ---
def log_validation(text_encoder, tokenizer, unet, vae, args, accelerator, weight_dtype, epoch):
    logger.info("Running validation...")
    pipeline = StableDiffusionPipeline.from_pretrained(
        args.pretrained_model_name_or_path,
        text_encoder=accelerator.unwrap_model(text_encoder),
        tokenizer=tokenizer,
        unet=accelerator.unwrap_model(unet),
        vae=vae,
        safety_checker=None,
        torch_dtype=weight_dtype,
    )
    pipeline.set_progress_bar_config(disable=True)
    pipeline = pipeline.to(accelerator.device)
    
    images = []
    generator = torch.Generator(device=accelerator.device).manual_seed(args.seed) if args.seed else None
    
    for _ in range(args.num_validation_images):
        with torch.autocast("cuda"):
            image = pipeline(args.validation_prompt, num_inference_steps=30, generator=generator).images[0]
            images.append(image)
            
    del pipeline
    torch.cuda.empty_cache()
    return images
```

```python
# --- 4. CONFIGURATION ---
@dataclass
class LoRAConfig:
    pretrained_model_name_or_path: str = "/kaggle/working/SD15_Baked_Gau"
    train_data_dir: str = "/kaggle/input/bovagau-poses/images/Gau"
    output_dir: str = "/kaggle/working/bo_lora_output"
    placeholder_token: str = "<Gau>"
    resolution: int = 512
    center_crop: bool=True
    flip_p: float=0.5
    train_batch_size: int = 4
    gradient_accumulation_steps: int = 4
    validation_prompt: str = "a photo of <Gau> wearing a vest suit"
    num_validation_images: int = 4
    validation_steps: int = 250
    learning_rate: float = 2e-5
    rank: int = 32
    max_train_steps: int = 3000 # Increase this for better results (e.g., 2000-3000)
    lr_scheduler: str = "constant_with_warmup"
    lr_warmup_steps: int = 200
    adam_weight_decay: float = 1e-3
    max_grad_norm=1.0,
    mixed_precision: str = "fp16"
    save_steps: int = 250
    repeats: int = 10
    seed: int = 36 # Added seed field
    enable_xformers_memory_efficient_attention: bool = True # Added optimization flag
    report_to: str = "wandb" 
    wandb_project_name: str = "lora-training"

    resume_from_checkpoint: str = None 
```

```python
config = LoRAConfig(
    resume_from_checkpoint="/kaggle/input/bovagau-lora/pytorch/default/10/Gau_4/Gau_4",
    max_train_steps=3000,
)
trainer = LoRATrainer(config)
trainer.train()
```
